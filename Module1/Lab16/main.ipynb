####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python

import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup

# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'

earthquake_list = [l for l in soup.find_all("tr", class_ = "normal", limit = 20)]

values = [m[0].text for m in [l.find_all("td", class_ = "tabev1") for l in earthquake_list]]
latitude = [m[0].text for m in [l.find_all("td", class_ = "tabev2") for l in earthquake_list]]
latitude_list = [values[i] + " " + latitude[i] for i in range(len(values))]

longitude = [m[1].text for m in [l.find_all("td", class_ = "tabev1") for l in earthquake_list]]
longitude_list = [m[1].text for m in [l.find_all("td", class_ = "tabev2") for l in earthquake_list]]
longs = [longitude[i] + " " + longitude_list[i] for i in range(len(longitude))]

data = {
    "date": [m.a.text[:m.a.text.find("   ")] for m in [l.find("td", class_ = "tabev6") for l in earthquake_list]],
    "time": [m.a.text[m.a.text.find("   "):] for m in [l.find("td", class_ = "tabev6") for l in earthquake_list]],
    "latitude": latitude_list,
    "longitude": longs,
    "region": [m.text for m in [l.find("td", class_ = "tb_region") for l in earthquake_list]] 
}

df = pd.DataFrame(data)
df
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'

html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
# Once we have our soup, we need to extract the table containing the passwords.
table = soup.find_all('table',{'class':'wikitable'})[0]
table

#We now have the table we want, but to be able to load the data into Pandas, we need to extract each of the rows (tags) and their cell values into a a nested list.
rows = table.find_all('tr', limit=10)
rows = [row.text.strip().split("\n") for row in rows]
rows

# From this nested list, we can specify what the column names are and then use the rest of the data to populate a data frame.
colnames = rows[0]
data = rows[1:]
 
df = pd.DataFrame(data, columns=colnames)
df
````
