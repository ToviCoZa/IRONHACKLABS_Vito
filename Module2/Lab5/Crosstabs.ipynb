### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(columns=moovies.country.count(), index=moovies['country'])
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
my_genres=moovies["genres"].values
pd.crosstab([moovies.director_name, moovies.genres, moovies.director_name],moovies.country)
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(moovies.color, moovies.duration).agg("mean")
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
a = [moovies.actor_1_facebook_likes + moovies.actor_2_facebook_likes + moovies.actor_3_facebook_likes + moovies.director_facebook_likes + moovies.movie_facebook_likes]

pd.crosstab(a, moovies.director_name).agg('mean')
```
